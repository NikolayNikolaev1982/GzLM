## 1. COSTRUZIONE DI UN MODELLO STATISTICO
I modelli esplicitano la relazione statistica e matematica fra variabili in modo semplificato secondo un trade off tra adattamento ai dati e parsimonia nel numero di parametri, perciò secondo l’assunto di Popper ogni modello è intrinsecamente sbagliato. La differenza sostanziale tra un modello Matematico e un modello Statistico consiste nel fatto che:
- Un modello matematico è deterministico e approssima in maniera esatta i dati, quindi la funzione che esplicita la relazione tra le variabili può essere così sintetizzata considerando un vettore y di variabili risposta e dei vettori di covariate individuali $y=f(x_1,x_@...x_k)$
- Un modello statistico è caratterizzato dalla presenza di errori all’interno della relazione esplicitata che dipendono da
  - Variazioni Individuali;
  - Componenti sistematiche non incluse nel modello;
  - Errori di misura;
  - Nel caso stocastico in cui gli errori dipendono dalla differenza tra campione e popolazione
la relazione tra le variabili può essere sintetizzata aggiungendo una componente stocastica $u$ all’equazione precedente ovvero: $y=(x_1x_2..x_k)+u$

Le fasi per procedere alla stima di un modello partono da considerazioni teoriche, che stanno alla base dell’ipotesi che vogliamo verificare, passando dalla verifica pratica di quanto ipotizzato

## 1.1. Specificazione di un modello:
Per poter definire un modello di regressione è importante definire:
- La variabile dipendente $(y)$, cioè la variabile che si vuole “spiegare” o il vettore di variabili risposta;
- Le variabili esplicative $(x_1,x_2, …, x_k)$, o i vettori delle covariate cioè le variabili disponibili e che si ritiene possano spiegare il fenomeno (covariate, regressori, variabili indipendenti) che rappresentano la componente sistematica ed osservata del modello con natura deterministica (non sono variabili casuali).
- La forma funzionale $f(.)$, cioè la relazione fissata a priori che lega la variabile dipendente alle variabili esplicative.

ll modello è specificato in riferimento alle unità statistiche (u.s.) di una data popolazione dalla quale si è estratto un campione di numerosità n, viene di solito rappresentato nella seguente maniera:

$$y_i=X_i\beta+ \epsilon_i \text \tag{1}$$
- i=1,...,n

Dove
- $y_i$ rappresenta la variabile dipendente osservata per l’unità $i$-esima;
- $X_i$ rappresenta la matrice delle variabili esplicative osservate per l’unità i-esima;
- $ε_i$ rappresenta l’errore per l’unità i-esima.

All’interno di questa specificazione, la componente non nota da stimare è rappresentata dal vettore $β$ (indicato in seguito anche con b) che rappresenta il vettore dei parametri ignoti (coefficienti di regressione), mentre la componente di errore $(ε_i)$ non è stimabile e resta ignota. La variabile errore , indicata in seguito anche con e solitamente stocastica sintetizza la nostra “ignoranza” circa la relazione che lega la risposta alle covariate (omissione di variabili esplicative, errori di misurazione, errori individuali).

La fase di verifica di un modello statistico è basata su una serie di:
- indicatori che misurano la bontà di adattamento di un modello rispetto ai dati;
- decisioni inferenziali (base su test statistici) che descrivono la significatività del modello e dei parametri (come, ad esempio, gli intervalli di confidenza) e le analisi dei residui del modello di regressione che consentono di valutare la capacità esplicativa del modello e la sua bontà di adattamento ai dati.

Non sempre la specificazione del modello è la migliore che si possa identificare; pertanto, se la fase di verifica non conduce al rifiuto del modello stimato allora il modello stimato può essere utilizzato, altrimenti si torna alla fase di specificazione.

I modelli che possono essere specificati sono molteplici e variano in base:
- alla variabile dipendente;
- alle variabili esplicative;
- al tipo di associazione (che può essere lineare oppure non lineare)

Una sintesi delle diverse opzioni è la seguente:
- Regressione lineare semplice: Una sola variabile dipendente e una sola variabile esplicativa
$$y=b_0b_1x_1+e \tag{2}$$


Regressione lineare multipla: Una sola variabile dipendente e più di una variabile esplicativa, rappresentato anche in forma estesa nella figura sottostante  
$$y=b_0+b_1x_1+...+b_kx_k+...+u=X'B+e \tag{3}$$

Regressione lineare multivariata: Più di una variabile dipendente e più di una variabile esplicativa
$$Y=XB+E\tag{4}$$

dove:
- $Y(n*r)=X_{n*p}B_{p*r}+E_{n*r}$ con Y matrice delle osservazioni per le variabili risposta;
- X matrice riferita all’insieme delle osservazioni delle r variabioli esplicative;
- B vettore dei parametri ignoti da stimare;
- E matrice degli errori casuali non osservabili.




$$
\mathbf{y} =
\begin{bmatrix}
y_1 \\
\vdots \\
y_n
\end{bmatrix}
\quad
\text{vettore delle } n \text{ osservazioni sulla variabile dipendente}
$$

$$
\mathbf{X} =
\begin{bmatrix}
1 & x_{11} & x_{12} & \cdots & x_{1k} \\
1 & x_{21} & x_{22} & \cdots & x_{2k} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_{n1} & x_{n2} & \cdots & x_{nk}
\end{bmatrix}
\quad
\text{Matrice costituita dall'insieme ordinato delle } n \text{ osservazioni sulle variabili esplicative}
$$

$$
\mathbf{b} =
\begin{bmatrix}
b_0 \\
b_1 \\
\vdots \\
b_k
\end{bmatrix}
\quad
\text{Vettore dei parametri ignoti da stimare}
$$

$$
\mathbf{e} =
\begin{bmatrix}
e_1 \\
\vdots \\
e_n
\end{bmatrix}
\quad
\text{vettore degli errori casuali ignoti da stimare}
$$


Se consideriamo solo due regressori all’interno del modello l’equzione diventa:

$$y_i=b_0+b_1x_{1i}+b_{2}x_{2i}+e_i \tag{6}$$
- i=1...n

dove
- y è la variabile dipendente riferita alla i-esima unità statistica
- $x_1, x_2$ sono le due variabili indipendenti (regressori)
- $y_{1i}, x_{1i}, x_{2i}$ sono riferite all’unità ’i-esima
- $b_0$ è un parametro del modello definito intercetta
- $b_1$ è il parametro che rappresenta l’effetto su y di una variazione unitaria in $x_1$, tenendo $x_2$ costante
- $b_2$ è il parametro che rappresenta l’effetto su y di una variazione unitaria in $x_2$, tenendo $x_1$ costante
- $e_i$ è il termine di errore ignoto riferito all’unità i-esima.


Esempio: interpretazione dei coefficienti di regressione nel modello di regressione multipla.
Si consideri un delta $Δx_1$ ottenuto facendo variare $x_1$ di tenendo $x_2$ costante. La retta di regressione per la popolazione di riferimento prima della variazione è definita come

$$y = b_{10} + b_{11}x_1 + b_{12}x_2 + e
 \tag{7}$$
La retta di regressione della popolazione dopo la variazione diventa:

$y + Δy = b_{10} + b_1(x_1 + Δx_1) + b_{12}x_2 + e,	\tag{8}$$
dove:
- $Δy = b_1 Δx_1$
- $b_1 = Δy / Δx_1$ tenendo $x_2$ costante
- $b_2 = Δy / Δx_2$ tenendo $x_1$ costante
- $b_0$ = valore atteso della risposta quando $x_1 = x_2 = 0$

## 1.2 Stima del modello, verifica delle ipotesi, e coeffcienti standardizzati

Per poter stimare i coefficienti di regressione, uno degli approcci utilizzati nel contesto lineare è il metodo dei minimi quadrati lineari. Questo metodo formula il problema di stima come il problema di ricercare quel vettore b che rende minima la norma del vettore di scarti o errori e:

$$min \sum_i(y_i-x_i b)^2=min \sum_iu_i^2 = min e'e=min(y-X'b)'(y-X'b) \tag{9}$$

Lo stimatore ottenuto minimizza la somma dei quadrati degli scarti fra i valori osservati $y_i$ mediante la combinazione lineare degli elementi del vettore $b$ moltiplicati per i valori osservati $x$_$ della matrice delle variabili esplicative. Procedendo al calcolo del gradiente (derivate parziali) della funzione obiettivo rispetto al vettore b si ottiene:

$$\frac{∂ e'e}{∂}= -2X'(y-Xb) \tag{10}$$

La condizione per l’esistenza di un minimo impone la nullità del gradiente:

$$X'(y-Xb)=0$$

che conduce al sistema di equazioni normali

$$X'Xb=X'y \tag{12}$$

Quando la matrice dei coefficienti sotto condizioni generali è di rango pieno, il sistema possiede un’unica soluzione che è rappresentata dalla seguente equazione:

$$\hat{b}=(X'X)^{-1}X'y \tag{13}$$

Lo stimatore rappresenta quindi lo stimatore dei minimi quadrati del vettore dei parametri. Ogni parametro $b_{1j} indica la variazione attesa per la risposta al variare unitario di $x_j$ mentre rimangono invariate le altre covariate in X.

È opportuno evidenziare come l’espressione (13) dello stimatore dei minimi quadrati coincida con lo stimatore lineare efficiente dei parametri e quindi porti a stimare efficientemente i parametri.

Se $D_x$ è la matrice diagonale i cui elementi diagonali sono le varianze delle variabili in X e si considerano standardizzare sia le variabili in X che le y (denotate con X* e y* nel seguito) cioè divise per il loro scarto quadratico medio allora il coefficiente standardizzato è il coefficiente di regressione fra X* e y*, ovvero

$$B^*=(X^{*'}X^*)^{-1}X^{*'}y^* \tag{14}$$

che misura l’associazione tra X e y al netto dell’ordine di grandezza con cui sono misurate le variabili.

È necessario valutare se e quanto questa relazione ipotizzata per il modello sia effettivamente plausibile alla luce dei dati osservati. Per fare ciò si valuta la bontà di adattamento del modello. Solitamente si usa l’indicatore denominato R2 che offre una misura quantitativa e relativa di adattamento.

Ogni volta che si specifica un modello, l’obiettivo è quello di trovare una relazione generale che permette di spiegare la variabilità del fenomeno di interesse e pertanto la variabilità della variabile risposta (y). Quest’ultima può essere decomposta in due parti: variabilità che il modello riesce a speigare e quella che il modello non è in grado di spiegare. Ricordandosi della (9) e successive ed esprimendo per semplicità y e X in termini di variabili centrate sulla media si ha:

$$\hat{y}=X\hat{b}+\hat{e}=\hat{y}+\hat{e} \tag{15}$$

dove:
- $\hat{y}= X(X'X)^{-1}X'y$   $(16)$
- $\hat{e}= (I-X(X'X)^{-1}X'y)$   $(17)$
- Devianza totale (total sum of squares, TSS): $TSS=y'y$  (18)
- Devianza spiegata (model sum of squares, MSS): rappresenta la somma dei quadrati degli scarti tra valori interpolati (y) e la media delle variabili indipendenti, ovvero somma dei quadrati dovuti alla regressione: $MSS = y’X(X’X)^{-1}Xy$. (19)
- Devianza residual (sum of squared residuals, SSR): $SSR=\hat{e'}\hat{e}=y'(I-X(X'X)^{-1}X')y$, 	(20) rappresenta la somma dei quadrati degli scarti tra valori osservati e interpolati (somma media degli scarti)

Poichè è dimostrabile che $\hat{y'}e=0$ (21), la scomposizione delle devianze per valutare la bontà di adattamento è la seguente:
$$TSS = MSS + SSR,	\tag{22}$$

con indice di adattamento definito come rapporto:

$$R^2 = MSS / TSS,  		\tag{23}$$

che prende il nome di coefficiente di correlazione multipla o di determinazione, e rappresenta la frazione della varianza di y spiegata dalle covariate presenti in X. L’indice $R^2$ varia tra 0 e 1 e aumenta sempre quando si aggiunge un altro regressore o variabile esplicativa, questo rappresenta un problema che può essere facilmente corretto con un termine di penalizzazione per evitare che il modello risulti sovra-parameterizzato.

## 2. Modello Lineare Clascio: Ipotesi:

Nella specificazione di un modello lineare l’associazione non è deterministica ma dipendente da variazioni casuali del termine di errore ($e_i$). La determinazione campionaria della variabile casuale non osservabile del camponente stocastica termine di errore quantifica l’errore che si commette al variare del campione spiegando la variabile y tramite le covariate in X.

Sia
$$y=Xb+e \tag{24}$$


dove:
- y rappresenta il vettore della variabile casuale - risposta;
- X rappresenta la matrice costituita dall’insieme ordinato delle n osservazioni campionarie sulle r variabili esplicative;
- b rappresenta  il vettore dei parametri ignoti del modello da stimare;
- e rappresenta il vettore degli errori casuali non osservabili.

L’equazione (24) con le assunzioni su ei definisce il modello di regressione multipla.

Nella specificazione del modello lineare classico, dobbiamo tenere in considerazione le seguenti assunzioni, senza la validità delle quali, il modello specificato potrebbe risultare non correttamente specificato:
- Linearità;
- Non sistematicità degli errori;
- Sfericità degli errori;
- Non stocasticità delle variabili esplicative;
- Non collinearità delle variabili esplicative;
Numerosità della popolazione.

A queste sei assunzioni, se ne aggiunge un’altra relativa agli stimatori e che permette di costruire test e intervalli di confidenza ovvero
- Distribuzione normale delle componenti di errore: $ε_i ∼ N(0,σ^2)$.



### 2.1 Linearità e additività
Questa ipotesi è riferita alla forma funzionale del modello. Un modello di regressione è lineare quando tutti i termini nel modello sono o la costante o un parametro moltiplicato per una variabile indipendente. Data la seguente specifica del modello, $y = b_0 + b_1x_1 + b_2x_2 + … + b_kx_k + e$ dove i parametri sono le stime a minimi quadrati e l'errore è casuale, la caratteristica che definisce la regressione lineare è questa forma funzionale dei parametri ed il fatto che i termini sono additivi. I modelli lineari possono modellare la curvatura includendo variabili non lineari come polinomi e trasformando funzioni esponenziali.

## 2.2 Non sistematicità degli errori
Il termine di errore rappresenta la variabilità della variabile dipendente che non viene spiegata dalle variabili indipendenti inserite nel modello. La casualità dovrebbe determinare i valori del termine di errore. Affinché il modello sia correttamente specificato, il valore medio del termine di errore deve essere uguale a zero.

Si consideri il modello scritto per la i-esima unità:
$$y_i = X_i b + e_i ,	\tag{25}$$

e si ripetano le osservazioni su diverse popolazioni o campioni. Ad esempio, si consideri un esperimento di laboratorio per studiare l’associazione tra il peso di un oggetto e la velocità della sua caduta ripetendo l’esperimento per n volte. Si consideri un numero k molto elevato di laboratori in ognuno di quali si ripete n volte lo stesso esperimento. Si considerino i risultati del primo esperimento in ogni laboratorio definiti come $e_{11}, e_{12}, …, e_{1k}$. Nel loro complesso tali risultati possono definirsi come realizzazioni da una variabile casuale $E_1$. Si può quindi calcolare il valore atteso e la varianza della variabile casuale $E_1$. Si considerino i risultati del secondo esperimento in ogni laboratorio $e_{21}, e_{22}, …, e_{2k}$. Nel loro complesso tali risultati possono definirsi come una seconda variabile casuale $E_2$ dalla stessa distribuzione di $E_1$. Analogamente avviene per i risultati i risultati dell’j-esimo esperimento in ogni laboratorio. Si hanno quindi nel complesso n variabili casuali che rappresentano gli esperimenti $E_1, E_2, …, E_i, …, E_n$ con identica distribuzione. Si considera il valore atteso e la varianza di ognuna di esse e la correlazione fra variabili casuali diverse $E_i$ e $E_j$. Per ciò che concerne il valore atteso nell’ipotesi di non sistematicità degli errori

$$E[E_i]=0 \tag{26}$$

e quinci

$$E[E|X]= 0 \tag{27}$$

e consegutemente:

$$E[y|X]=Xb \tag{28}$$

$$E[y]=E[Xb+e]= E[Xb]+E[E]=Xb \tag{26}$$

Se l'errore medio fosse diverso da zero questo indicherebbe che il nostro modello sottostima sistematicamente tutti i valori osservati. Questo implicherebbe che il modello stimato sia inadeguato perché non è corretto in media.

Pertanto, nella stima di un modello si ipotizza che che il valore atteso del termine di errore sia uguale a zero. Altrimenti, se il valore atteso fosse +7 anziché zero, la parte del termine di errore sarebbe prevedibile e dovremmo aggiungere queste informazioni al modello di regressione stesso, mentre la specificazione del modello implica che solo gli errori casuali rimasti finiscano all’interno del termine dell'errore.

Di questa ipotesi, non è necessario preoccuparsi quando si include la costante nel modello di regressione perché si impone che la media dei residui sia uguale zero. Inoltre il termine di errore e le covariate debbono essere indipendenti.

## 2.3 Sfericità degli errori (omoschedastici e non correlati)

La varianza degli errori dovrebbe essere coerente per tutte le osservazioni. In altre parole, la varianza degli errori non cambia per ogni osservazione o per un intervallo di osservazioni. Questa condizione è nota come omoschedasticità (stessa dispersione). Se la varianza cambia, allora siamo in una situazione di eteroschedasticità (dispersione diversa). Occorre ricordare che il modello lineare

$$y=Xb+e \tag{30}$$

con $y_{n*1}, X_{n*k}, b_{k*1}, e_{n*1}$  presenta la seguente variabilità

$$\sigma^2_y= b'\sum_Xb + \sigma^2 \tag31{}$$
- con $\sigma^2= e'e$

Ma considerando il modello $y_i = X_i b + e_i$  al variare di popolazioni o campioni si generano n variabili casuali $E_1, E_2, …, E_i, …, E_n$ dalla identica distribuzione. Si possono calcolare oltre al valore atteso le n varianze di ogni variabile casuale $Var(E_i)$ e le $(n^2-n)$ correlazioni fra variabili casuali diverse $E_i$ e $E_j$.

Data la (30) si ha:
$$Var[E_i]= E[E_i^2]=\sigma_i^2 \tag{31}$$

$$Cor[E_i, E_j]= \frac{E_iE_j'}{n}= \rho_{ij} \tag{32}$$

Se gli errori sono omoschedastici si ha $\sigma_1^2=\sigma_2^2=...=\sigma_n^2$ ovvero caratterizzati da una variabilità che non dipende dall’istante di rilevazione del fenomeno o dal tipo di unità statistica, in questo caso dal tipo di laboratorio.
- Omoschedasticità: campioni di esperimenti fatti in serie in laboratorio e tenuti nelle stesse condizioni in riferimento a una variabile dipendente quale la velocità di caduta di un oggetto e variabile esplicativa quale il peso dell’oggetto stesso. La parte della varianza non spiegata dalla variabile esplicativa per l’esperimento 1, al variare del campione è pari a quella degli altri esperimenti 2, 3, 4, …, n.
- Omoschedasticità forzata: campioni di unità in cui si studia il nesso tra peso e altezza. La parte della varianza non spiegata dalla variabile esplicativa, al variare del campione dell’esperimento 1 non può essere pari a quella degli altri esperimenti 2, 3, 4, … , n perché il nesso peso altezza varia da individuo a individuo.

Se gli errori sono incorrelati si ha $\rho_{12}=\rho_{13}=...=\rho_{nn}= 0$ gli errori sono senza legami stocastici apparenti.

- Incorrelazione: Si consideri ad esempio l’associazione tra il voto e l’aiuto allo studio per studenti residenti in diverse città. Al variare del campione la parte della varianza non spiegata dalle variabili esplicative degli studenti nella città 1 non può essere legata a quella degli studenti nelle città 2,3,4,...
- Incorrelazione forzata: Si considerino delle rilevazioni su un campione di dati ripetute nel tempo (si tratta di dati longitudinali (o panel) in cui si dipone di più rilevazioni per la stessa unità statistica) con cui si intende analizzare l’associazione tra l’acquisto di macchine in anni diversi ed il reddito. Il fatto di averla acquistata nell’anno 1 influenza il fatto di acquistarla negli anni successivi.

Se gli errori sono omoschedastici (con uguale varianza) ed incorrelati si ha:

$$E[e]= E[y-Xb]= (Xb-Xb)= 0 \tag{33}$$

E quindi:

$$ Var[e_i]= E[e_i^2]=\sigma^2 \tag{34}$$

$$Cov(e_i, e_j)= E(e_j)=0 \tag{35}$$

Quindi la matrice di varianze-covarianze degli errori è di tipo scalare.


Il modo più semplice per verificare questa ipotesi è quello di rappresentare tramite diagramma a dispersione i valori dei residui rispetto ai valori interpolati (ovvero i valori delle risposta previsti dal modello). Se i residui non sono equamente distribuiti nel piano cartesiano, e il grafico mostra una forma a cono (si veda il seguente grafico) in cui la varianza dei residui aumenta in una certa direzione, allora siamo in presenza di eteroschedasticità che comporta una riduzione della precisione delle stime nella regressione lineare OLS.



```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Simulazione dati con eteroscedasticità
np.random.seed(1)
X = np.random.rand(100, 1) * 70
noise = np.random.randn(100) * (X.ravel() / 10)  # varianza cresce con X
y = 5 + 0.5 * X.ravel() + noise

# Fit modello di regressione
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

# Grafico residui vs valori predetti
plt.figure(figsize=(8, 5))
plt.scatter(y_pred, residuals)
plt.axhline(0, color='gray', linestyle='--')
plt.title('Versus Fits\n(response is Accidents)')
plt.xlabel('Fitted Value')
plt.ylabel('Residual')
plt.grid(False)
plt.show()
```

Nel caso in cui l'errore di un'osservazione sia positivo e aumenti sistematicamente la probabilità che l'errore riferito alla successiva unità sia positivo, siamo in presenza di una correlazione positiva, mentre nel caso in cui l'errore riferito ad un’unità successiva abbia una più alta probabilità di avere il segno opposto, siamo in presenza di una correlazione negativa. Questo problema è noto sia come correlazione seriale che come autocorrelazione.

## 2.4 Non stocasticità delle variabili esplicative
Se una variabile indipendente fosse correlata al termine d'errore, questo ci impedirebbe di usare la variabile indipendente per prevedere il termine d'errore, andando così a violare l’assunzione che il termine di errore sia un errore casuale imprevedibile. Dobbiamo trovare un modo per incorporare tali informazioni nel modello di regressione stesso. Questa ipotesi viene anche definita come esogeneità. Quando esiste questo tipo di associazione, c'è l'endogeneità.

Le violazioni di questa ipotesi possono verificarsi perché vi è simultaneità tra le variabili indipendenti e dipendenti, la deviazione della variabile omessa o l'errore di misurazione nelle variabili indipendenti. La violazione di questa ipotesi condiziona la stima del coefficiente.

Formalmente sia $x_j$  la j-esima covariata considerata fissa per j = 1, 2, … , p, ovvero si assume che  le determinazioni delle variabili esplicative non siano soggette a fluttuazioni stocastiche. Perciò:
$$E(X) = X		\tag{35}$$
ossia il valore atteso della matrice X delle covariate coincide con la matrice stessa. Inoltre:
$$cov(X, ε) = 0. 		\tag{37}$$

## 2.5 Non collinearità delle variabili esplicative
Premettendo che esiste perfetta correlazione lineare se due variabili hanno un coefficiente di correlazione di Pearson di +1 o -1, si ha che in questo caso quando una delle variabili cambia, anche l'altra variabile varia di una proporzione fissa; pertanto, le due variabili si muovono all'unisono. Ad esempio, la temperatura in gradi Fahrenheit è perfettamente correlata con la temperatura misurata in gradi Celsius (correlazione positiva con indice di Pearson di +1).

Il metodo di stima a minimi quadrati ordinari non può essere applicato quando alcune covariate sono perfettamente correlate. Questo dipende dal fatto che la matrice X non sarebbe a rango pieno, e quindi la matrice $X’X$ necessaria per il calcolo dei coefficienti sarebbe singolare e quindi non invertibile.  Formalmente se la matrice X ha rango pieno, pari a p+1 ovvero il numero delle covariate, includendo anche la costante
     $$ Rango (X) = p+1		\tag{38}$$
Ne consegue che la matrice $(X’X)$ non è singolare.

## 2.7 Normalità degli errori
Nel caso di un campionamento casuale semplice ogni osservazione è identicamente e indipendentemente distribuita per ogni i = 1, 2, …, n.

I minimi quadrati ordinari non richiedono che il termine di errore segua una distribuzione normale per produrre stime con varianza minima. Tuttavia, la soddisfazione di questa ipotesi consente di eseguire test statistici per valutare delle ipotesi sui parametri e ricavare intervalli di confidenza e intervalli di previsione affidabili.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Dati simulati (residui)
np.random.seed(42)
residuals = np.random.normal(0, 3, 40)
residuals[0] = -10.5
residuals[1] = 10.5

# Calcola i quantili teorici e osservati
osm, osr = stats.probplot(residuals, dist="norm", fit=False)

# Fit della linea rossa
slope, intercept, *_ = stats.linregress(osm, osr)
line_fit = slope * np.array(osm) + intercept

# Grafico stile Minitab
fig, ax = plt.subplots(figsize=(7.5, 4.5), facecolor="#E5E5E5")
ax.set_facecolor("#E5E5E5")

# Punti blu
ax.plot(osm, osr, 'o', color='navy', markersize=4)

# Linea di riferimento
ax.plot(osm, line_fit, color='darkred', linewidth=1)

# Titolo e assi
ax.set_title("Normal Probability Plot\n(response is %Fat)", fontsize=11)
ax.set_xlabel("Theoretical Quantiles", fontsize=9)
ax.set_ylabel("Residual", fontsize=9)

# Layout pulito
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simula residui
np.random.seed(42)
residuals = np.random.normal(0, 3, 40)
residuals[0] = -10.5
residuals[1] = 10.5

# Calcolo dei quantili teorici e residui ordinati
osm, osr = stats.probplot(residuals, dist="norm", fit=False)

# Calcolo dei percentili teorici
percentiles = stats.norm.cdf(osm) * 100  # da 0 a 100

# Regressione per linea rossa
slope, intercept, *_ = stats.linregress(percentiles, osr)
line_fit = slope * np.array(percentiles) + intercept

# Crea grafico stile Minitab
fig, ax = plt.subplots(figsize=(7.5, 4.5), facecolor="#E5E5E5")
ax.set_facecolor("#E5E5E5")

# Punti blu
ax.plot(percentiles, osr, 'o', color='navy', markersize=4)

# Linea rossa di riferimento
ax.plot(percentiles, line_fit, color='darkred', linewidth=1)

# Titolo e assi
ax.set_title("Normal Probability Plot\n(response is %Fat)", fontsize=11)
ax.set_xlabel("Percent", fontsize=9)
ax.set_ylabel("Residual", fontsize=9)

# Imposta tick dell'asse X manualmente
ax.set_xlim(0, 100)
ax.set_xticks([0, 1, 5, 10, 25, 50, 75, 90, 95, 99, 100])

# Aspetto grafico
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

Il modo più semplice è valutare se i residui di regressione possono essere assunti come determinazioni da una distribuzione normale. Se i residui seguono la linea retta sul grafico che vede i residui sull’asse delle ascisse ed i percentili della distribuzione normale standard sull’asse delle ordinate allora la loro distribizione empirica è assimilabile ad una distribuzione normale. Avendo assunto $ε_i  ∼ N(0,σ^2)$, ovvero che la distribuzione dell’errore campionario sia normale, ed essendo fisse le variabili in X risulta che anche la distribuzione di y è normale. Pertanto:
$$y ∼ N(X_b, σ^2I_n). 	\tag{39}$$
Quindi anche
$$\hat{B}=(X'X)^{-1}X'y \tag{40}$$

e' distribuito in modo normale. Quindi:

$$\hat{B}\sim N(b, \sigma^2(X'X)^{-1}) \tag{41}$$

Per n grande (maggiore di 25) per il teorema del limite centrale la distribuzione di $ε_i$ e quindi di y e $\hat{B}$ tendono alla normalità.
